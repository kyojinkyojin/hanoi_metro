```python
import geopandas as gpd
import matplotlib.pyplot as plt
from pyproj import Transformer
```

```python
metro = gpd.read_file("./geojson/Hanoi_Metro_Line_Official_8_666992285949592975.geojson", crs="EPSG:3857")
metro = metro.to_crs("EPSG:4326")
metro.to_file("./geojson/hn_metro.geojson", driver="geojson")
vn1 = gpd.read_file("./geojson/vn.json")
fig, ax = plt.subplots(figsize=(10,10))
vn1[vn1['name']=="Ha Noi"].plot(ax=ax, color="red", aspect=1)
metro.plot(ax=ax, aspect=1)
plt.show()
```
