```python
from shapely.ops import linemerge
import geopandas as gpd
from shapely import LineString, MultiPoint
```

```python
dfny = gpd.read_file("../nyc_metro/NYC_subway_lines.geojson")
dfny['name'] = dfny['name'].str.split("-")
dfny = dfny.explode("name")
df_g = dfny.groupby("name").agg({
    'geometry': lambda x: linemerge(list(x))
}).reset_index()
#t_df = dfs[0]
#t_df = t_df[['name', 'geometry']]

def merge_lines(lines):
    last = None
    points = []
    if isinstance(lines, LineString):
        return lines
    else:
        coords = [list(line.coords) for line in lines.geoms]
        outline = sorted([point for segment in coords for point in segment], key=lambda x: x[1], reverse=True)
        #print(outline)
        mpoints = MultiPoint(outline)
        #outline = LineString([i for sublist in outcoords for i in sublist])
    return outline

df_g['geometry'] = df_g['geometry'].apply(merge_lines)
#df_g = gpd.GeoDataFrame(df_g, geometry="geometry")
#df_g[df_g["name"] == "5"].plot()
#df_g.plot()
```

```python
from scipy.spatial import ConvexHull
import numpy as np

your_points = df_g[df_g["name"] == "5"].geometry.values
your_points = [point for segment in your_points for point in segment]
print(your_points)


# Step 1: Convex Hull to identify potential main branch
points = np.array([point for point in your_points])  # your_points is a list of tuples
hull = ConvexHull(points)

# Extract the longest segment
hull_lines = [(points[simplex, 0], points[simplex, 1]) for simplex in hull.simplices]
main_branch = max(hull_lines, key=lambda line: LineString(line).length)

# Step 2: Calculate distances to the main branch
main_line = LineString(main_branch)
branches = []
main_line.plot()
```
